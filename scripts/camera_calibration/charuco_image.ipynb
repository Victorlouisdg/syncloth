```python
from airo_camera_toolkit.calibration.fiducial_markers import AIRO_DEFAULT_CHARUCO_BOARD
import numpy as np
import matplotlib.pyplot as plt
```

```python
board = AIRO_DEFAULT_CHARUCO_BOARD
```

```python
pixels_per_cell = 100
margin_pixels = pixels_per_cell // 5

image_size = (7 * pixels_per_cell + margin_pixels, 5 * pixels_per_cell + margin_pixels)
image = np.zeros((image_size[1], image_size[0]), dtype=np.uint8)

image = board.generateImage(image_size, image, margin_pixels)
```

```python

plt.imshow(image, cmap='gray')
```
